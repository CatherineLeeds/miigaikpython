```python
text=input('Введите слово ')
if len(text) % 2 == 0:
    leng = len(text)//2
    sr = text[leng-1:leng+1]
    print("Средний символ", sr)
else:
    leng = len(text)
    sr = leng // 2
    print ("Средний символ:", text[sr])
```

Задача 6 Если бы мне платили каждый раз (3 балла)
Вам дан текст. Необходимо вычислить сколько раз встречается каждое слово в тексте, сколько различных слов было в тексте и сколько всего было слов. Затем замените количество появлений слова в тексте, на его частоту (количество встреч / общее количество слов в тексте). Найдите top 10 самых часто встречающихся слов длины > 3 и выпишите их.!!!!!!!!!!!!!

Ввод 
['my', 'world', 'my', 'rules']
Вывод
{'my':2, 'world':1, 'rules':1}
Различных слов - 3, всего слов - 4.
{'my':0.5, 'world':0.25, 'rules':0.25}

```python
text = input('Введите текст ') 
text=text.lower()
zam=[',','?','»','«','—']
for i in zam:
    text=text.replace(i,'')
kols={}
for s in text.split():
    kols.update({s:0})
for s in text.split():
    kols[s]=kols[s]+1
f=len(text.split())

print(kols)
print('Различных слов -',len(kols),'всего слов -',f)
kl={}
for k,v in kols.items():
    kl.update({k:v/f})

print('{',end='')
for k,z in kl.items():
    print(k,':',"%.3f" % z, end=', ')
print('}',end='')

sln=list(kl.values())
sln=sorted(sln)
sln.reverse()
top=[]
for i in sln:
    for w,k in kl.items():
        if len(top)<10 and k==i and len(w)>3 and w not in top:
            top.append(w)    
print()
print(*top)
    
```

Задача 7 Последовательные строки (3 балла)
На вход подается список, состоящий из строк, и целое число $k$. Ваша цель вывести на экран первую встретившуюся самую длинную строку, состоящую из k последовательных элементов списка. Если $k&lt;0$, список пуст или $k &gt;$ длины списка вывести пустую строку. 
Входные данные:list = [cat city exception dog exception kick], $k = 2$. Выходные данные: "cityexception"

```python
text=input("Введите текст ")
text=text.split(' ')
k=int(input('K= '))
if k<0 or k>len(text):
    print()
else:
    j=0
    g=0
    for i in range(0,len(text)-k+1):
        f="".join(text[i:k+i])
        if len(f)>g:
            j=i
            g=len(f)
    print("".join(text[j:k+j]))
```

Задача 9 "+1" 

```python
text=input("Введите строку: ")
alfa = "abcdefghijklmnopqrstuvwxyz"
alf = "абвгдеёжзийклмнопрстуфхцчшщъыьэюя"
chislo='0123456789'
if text[-1] in alfa or text[-1] in alf:
    text=text+'1'
else:
    k=0
    for i in text[::-1]:
        if i in chislo:
            k=k+1
        else:
            break
    text=text[:-k]+str(int(text[-k:])+1)        
print(text)
```
